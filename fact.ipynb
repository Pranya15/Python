```python
"""take a input from user if the no is even calculate even factorial and if odd calculate odd fact"""
def factors(n):
    for i in range(1,n+1):
        if n%i==0:
            print(i,end=" ")
factors(12)
```

```python
def even_fact(n):
    ans=1
    for i in range(2,n+1,2):
      ans*=i
    return ans
def odd_fact(n):
    ans=1
    for i in range(1,n+1,2):
        ans*=i
    return ans

n= int(input("Enter a no."))
if n%2==0:
    print(even_fact(n))
else:
    print(odd_fact(n))
```
