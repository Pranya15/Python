```python
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact


def nCr(n, r):
    if r > n:
        return 0
    
    return factorial(n) // (factorial(r) * factorial(n - r))


n = int(input("Enter n: "))
r = int(input("Enter r: "))

print("nCr =", nCr(n, r))
```
