```python
"""wap which will take int list from user in which even no will be inc by 5 and odd will dec by 5"""
lst=eval(input("Enter a list"))
for i in range(len(lst)):
    if lst[i]%2==0:
        lst[i]+=5
    else:
        lst[i]-=5
print(lst)
```
