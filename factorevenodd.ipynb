```python
"""wap to find even and odd factor count"""

n=int(input("Enter a no:"))
i=1
even_count=0
odd_count=0
while i<=n:
    if n%i==0:
        if i%2==0:
            even_count += 1
        else :
            odd_count += 1
    i +=1
print("Even factors:",even_count)
print("Odd factors:",odd_count)

```
